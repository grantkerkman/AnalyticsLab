# PyCitySchools Analysis

Two datasets are provided to analyze student test performance in math and reading of the PyCity Independent School District (PyISD). The ISD would like to examine the relationships between funding and student performance for this year's fiscal planning cycle. The ISD is under a lot of pressure to reduce costs so the city can divert funding to the I-35 expansion and hopefully finish the project by 2035. The analysis will provide recommendations on areas for further investigation by PyISD administration. <br> <br>

## PyCity ISD Summary
PyCity ISD has a current enrollment of 39,170 students and operates a mixed-type configuration with seven district schools and eight charter schools. The current budget for the ISD is $ 24,649,428. The average student math score is 79 with 75.0\% of the students passing. The average student reading score is 82 with 85.8\% of the students passing. Overall, 65.2\% of the students in the ISD are passing both tests. Passing is defined as a score greater than or equal to 70.

The top five performing schools (Cabrera, Thomas, Griffin, Wilson, and Pena) all achieved overall passing rates greater than 90\%, with average math scores of 83 and average reading scores of 84. All five schools are charter type and span all but the highest group of spending per student metrics. The five schools also cover the small, medium, and large designations for student population.

The bottom five performing schools (Rodriguez, Figueroa, Huang, Hernandez, and Johnson) have overall passing rates slightly above 50\%, with average math scores of 77 and average reading scores of 81. All five schools are district type, and all have greater than $635 spending per student metrics. The five schools have large enrollments, with three schools having 4,000 or more students and two of the schools having nearly 3,000. <br> <br>

## Performance as a Function of Funding
Conventional wisdom would point to a positive linear relationship of funding and performance, as one expects an investment to yield a positive return. PyISD testing data offers a reversal in conventional wisdom. Grouping the schools into four categories of spending per student, one can see that the percent passing overall drops sharply once a school spends $ 630-$645 per student. The schools spending the least per student (under $ 585) are performing the best with 90% passing both tests.

Republican rejoice, we can cut taxes and slash funding some more! But before this potential mistake is made, one should also acknowledge the datasets that are missing. Spending per student is a sufficient but not necessary metric to gauge performance. The data provided did not present information on faculty size or the number of 504 students enrolled (which could be used to determine the range of pay at a school, since specialists typically have higher salaries). Faculty size would point to the teacher/student ratio and shed light on the costs associated with school administrators, who like specialists have significantly higher salaries than teachers. Separating the payroll of instructors and administration allows for a better picture of the teacher-to-student ratio. <br> <br>

## Performance as a Function of School Size

PyCity has been the fastest-growing metro area in the country for nearly a decade, with no end in sight. Business-friendly tax cuts and lax regulations have seen corporations flock to the area. Like other areas of infrastructure, the city has been struggling to keep up with the YoY double-digit population increases. Looking at school performance as a function of school size will allow one to see the effects of expansion. Math scores see a small (6-7) point drop-off when enrollment is greater than 2000 students, with minimal differences in reading scores. On a percent passing basis, there is minimal difference in the math, reading, and overall test scores for schools with less than 2000 student enrollment. Once the jump to large schools is made, a different story plays out, with 58\% passing overall compared to the 90\% overall passing for schools under 2000 students.

This data suggests that overcrowding and large schools have a significant impact on the percent of students passing. Notice that the spread in scores is minimal, meaning low test scores are dragging the average scores down more significantly at large schools. Similar to the request for faculty size earlier, a teacher-to-student ratio would be extremely beneficial.

## Performance as a Function of School Type

At the heart of the current debate around education in this country, charter schools have emerged as the seemingly miraculous solution to both falling test scores and rising costs of education. The data from PyISD suggests this as well. Charter schools are drastically outperforming district schools, with the overall passing rate 46\% higher in charter schools. Test scores in math and reading for charter schools are 83 and 84, respectively, while district schools are scoring 77 and 81, respectively. There is a large discrepancy in the rate of students passing math and reading as well, with charter schools passing 94\% of students in math and 97\% of students in reading. This again stands in stark contrast to the district schools, with only 67\% of students passing math and 81\% passing reading.

Charter schools are the answer? Maybe, but there's probably more to it than that. Something curious to note is that only one charter school has an enrollment larger than 2000 students, while the smallest district school has an enrollment of 2700. We saw earlier that scores fall off the cliff at 2000+ enrollment, and it shouldn't surprise anyone that if almost all charter schools are small, then they must perform better. School type and size may be interchangeable in the current PyISD landscape.

## Conclusion
Based on the provided data, we can draw these conclusions:

* Funding plays less into student performance than conventional (business) wisdom would dictate. We found that the spending per student didn't have much of an impact on performance, as we had instances of schools covering nearly the entire range of funding performing well. The highest spending schools are some of the worst performing, meaning more money does actually equal more problems.

* School size is the KPI that needs attention. We are finding time and time again that the size of a school has the biggest impact on passing rates. Note that we see drops in test score (roughly 10\% from medium to large schools), but we see a nearly 25\% drop in passing rates of math. This drop is not as large in reading, indicating that the two subjects may need to be approached from different CI philosophies.

* Charter schools appear to be the answer, but there are still more questions. Charter schools span the range of spending per student but did not have many schools with large enrollment. The most expensive schools in PyISD are district schools with larger enrollments. Since there is only one charter school with larger than 2000 student enrollment, it is difficult to compare the PyISD district schools to charter schools simply because that would just be a comparison of small/medium enrollment to large enrollment performance.

What's missing from this analysis is the faculty employment numbers, ideally with years of service included. All PyISD employees are state government employees and have published and publicly available pay rates. By analyzing the payroll, one can determine what the teacher-to-student ratio is and how it differs from the faculty-to-student ratio. Administrators may or may not have a direct impact on student performance (do you really think the PyISD accounting team is playing a crucial role in the students learning Algebra II?) and the inclusion of their disproportionate salaries may be dragging the spend per student higher for the larger schools.

Charter schools may not be the correct solution for falling test scores, but it is the necessary reset that might be needed for education. With the melding of corporate values into our civic values in the 1980s, conventional wisdom held that economies of scale and business practices could transfer to all sorts of other industries. But a Fortune 500 company isn't a high school. Management cannot "fire" a student and replace them with another worker. The charter schools in PyISD indicate that school enrollment should be capped and new schools should be added.

Lastly, it should be noted that the focus on average scores should be taken into consideration with the standard deviation of scores. A quick analysis of the dataset yielded a significantly higher standard deviation in test scores for the poorly performing schools, indicating that a larger number of poor test results were skewing the data. It should be obvious at this point now that with the presence of such a large number of at-risk students, the only way to boost overall performance is to address these students' needs, and the first step is posting a job requisition for a math teacher.

```python
#Dependencies and Setup
import pandas as pd
from pathlib import Path

#Load and Read Files
school_data_csv = Path("Resources/schools_complete.csv")
student_data_csv = Path("Resources/students_complete.csv")

school_data = pd.read_csv(school_data_csv)
student_data = pd.read_csv(student_data_csv)

#Merge Datasets
school_data_complete = pd.merge(student_data, school_data, how="left", on =["school_name"])
school_data_complete.head()
```

## District Summary

```python
#Total Number of Unique Schools
school_count = len(school_data_complete["School ID"].unique())

school_count
```

```python
#Total Number of Students
student_count = len(school_data_complete["Student ID"].unique())

student_count
```

```python
#Total Budget ($)
total_budget = school_data["budget"].sum()

total_budget
```

```python
#Average Math Score
average_math_score = school_data_complete["math_score"].mean()

average_math_score
```

```python
#Average Reading Score
average_reading_score = school_data_complete["reading_score"].mean()

average_reading_score
```

```python
#Passing Math Percentage (Grade >= 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / student_count * 100

passing_math_percentage
```

```python
#Passing Reading Percentage (Grade >= 70)
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / student_count * 100

passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
#Disctrict Summary Table
district_summary_df = pd.DataFrame({
    "Total Schools": [school_count],
    "Total Students": [student_count],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [passing_math_percentage],
    "% Passing Reading": [passing_reading_percentage],
    "% Overall Passing": [overall_passing_rate]
})

district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df
```

## School Summary

```python
#Select the School Types in the District
school_types = school_data_complete.groupby("school_name")["type"].unique().str.join(', ')

school_types
```

```python
#School Enrollment per school
total_students_per_school = school_data_complete.groupby("school_name")["size"].count()

total_students_per_school
```

```python
#Convert the size and budget to integers
school_data = school_data.astype({"School ID": object})

school_data.dtypes
```

```python
#Calculate the Total Budget per school
per_school_budget = school_data_complete.groupby("school_name")["budget"].mean()

per_school_budget
```

```python
#Calculate the per capita spending for each school
school_data_complete["per_student_budget"] = (school_data_complete["budget"] / school_data_complete["size"])
per_capita = school_data_complete.groupby("school_name")["per_student_budget"].mean()

per_capita
```

```python
#Calculate the average math scores for each school
per_school_math = school_data_complete.groupby("school_name")["math_score"].mean()

per_school_math
```

```python
#Calculate the average reading scores for each school
per_school_reading = school_data_complete.groupby("school_name")["reading_score"].mean()

per_school_reading
```

```python
#Percent of students passing Math per school
per_school_math_passing = school_data_complete[school_data_complete["math_score"] >= 70].groupby("school_name").count()["student_name"] / (total_students_per_school) * 100

per_school_math_passing
```

```python
#Percent of students passing Reading per school
per_school_reading_passing = school_data_complete[school_data_complete["reading_score"] >= 70].groupby("school_name").count()["student_name"] / (total_students_per_school) * 100

per_school_reading_passing
```

```python
#Percent of students passing both subjects
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size() / (total_students_per_school) * 100

school_students_passing_math_and_reading
```

```python
#Create the summary DataFrame for the School Summary
per_school_summary = pd.DataFrame({
    "School Type": school_types,
    "Total Students": total_students_per_school,
    "Total School Budget": per_school_budget,
    "Per Student Budget": per_capita,
    "Average Math Score": per_school_math,
    "Average Reading Score": per_school_reading,
    "% Passing Math": per_school_math_passing,
    "% Passing Reading": per_school_reading_passing,
    "% Overall Passing": school_students_passing_math_and_reading
})

#Format the table
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"].map("${:,.2f}".format)

per_school_summary
```

## Highest Performing Schools (by % Overall Passing)

```python
#Sort schools in decending order by % oveall Passing
top_schools = per_school_summary.sort_values("% Overall Passing", ascending = False)

top_schools.head(5)
```

## Math Scores by Grade

```python
#Create lists for each of the four grade levels (9th - 12th)
ninth_graders = school_data_complete[school_data_complete["grade"] == "9th"]
tenth_graders = school_data_complete[school_data_complete["grade"] == "10th"]
eleventh_graders = school_data_complete[school_data_complete["grade"] == "11th"]
twelfth_graders = school_data_complete[school_data_complete["grade"] == "12th"]

#Group by `school_name` and take the mean of the `math_score` column for each grade level
ninth_grade_math_scores = ninth_graders.groupby("school_name")["math_score"].mean()
tenth_grade_math_scores = tenth_graders.groupby("school_name")["math_score"].mean()
eleventh_grade_math_scores = eleventh_graders.groupby("school_name")["math_score"].mean()
twelfth_grade_math_scores = twelfth_graders.groupby("school_name")["math_score"].mean()

#Combine each of the scores above into a single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({
    "9th": ninth_grade_math_scores,
    "10th": tenth_grade_math_scores,
    "11th": eleventh_grade_math_scores,
    "12th": twelfth_grade_math_scores  
})

#Minor data wrangling
math_scores_by_grade.index.name = None

#Display the DataFrame
math_scores_by_grade
```

## Reading Scores by Grade

```python
#Create lists for each of the four grade levels (9th - 12th)
ninth_graders = school_data_complete[school_data_complete["grade"] == "9th"]
tenth_graders = school_data_complete[school_data_complete["grade"] == "10th"]
eleventh_graders = school_data_complete[school_data_complete["grade"] == "11th"]
twelfth_graders = school_data_complete[school_data_complete["grade"] == "12th"]

#Group by `school_name` and take the mean of the `math_score` column for each grade level
ninth_grade_reading_scores = ninth_graders.groupby("school_name")["reading_score"].mean()
tenth_grade_reading_scores = tenth_graders.groupby("school_name")["reading_score"].mean()
eleventh_grade_reading_scores = eleventh_graders.groupby("school_name")["reading_score"].mean()
twelfth_grade_reading_scores = twelfth_graders.groupby("school_name")["reading_score"].mean()

#Combine each of the scores above into a single DataFrame called `math_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({
    "9th": ninth_grade_reading_scores,
    "10th": tenth_grade_reading_scores,
    "11th": eleventh_grade_reading_scores,
    "12th": twelfth_grade_reading_scores  
})

#Minor data wrangling
reading_scores_by_grade.index.name = None

#Display the DataFrame
reading_scores_by_grade
```

## Scores by School Spending

```python
#Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
#Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()

school_spending_df
```

```python
#Convert "Per Student Budget" column to string and remove the non-numeric values we added earlier
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].astype(str)
school_spending_df["Per Student Budget"] = school_spending_df["Per Student Budget"].str.replace("$", "").str.replace(",", "")

#Now we can convert "Per Student Budget" column to numeric to resolve the int-str error in our bins
school_spending_df["Per Student Budget"] = pd.to_numeric(school_spending_df["Per Student Budget"])

#Apply pd.cut() to categorize spending based on the bins
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(
    school_spending_df["Per Student Budget"],
    spending_bins,
    labels=labels,
    include_lowest=True
)

school_spending_df
```

```python
#Calculate averages for the desired columns
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
```

```python
#Assemble into DataFrame
spending_summary = pd.DataFrame({
    "Math Scores (avg)": spending_math_scores,
    "Reading Scores (avg)": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Passing Overall": overall_passing_spending    
}) 

#Display results
spending_summary
```
